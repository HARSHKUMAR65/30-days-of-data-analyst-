# Data_cleaning

```python
data['Subscribers']=data['Subscribers'].str.replace(',','')
data['Video Views']=data['Video Views'].str.replace(',','')
```

```python
data['Subscribers']=data['Subscribers'].astype('int64')
data['Video Views']=data['Video Views'].astype('int64')
```
