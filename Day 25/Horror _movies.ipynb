```python
##import libraries 
import pandas  as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
import plotly.express as s
from wordcloud import WordCloud,STOPWORDS
```

```python
## read data 
data = pd.read_csv("horror_movies.csv")

data.head()
```

```python
##cheaking shape 
data.shape
```

```python
data.info()
```

```python
##cheaking null value 
data.isnull().sum()
```

```python
data.fillna(0,inplace=True)
data['release_date']=pd.to_datetime(data['release_date'])
data['year']=data['release_date'].dt.year
```

# EDA

```python
data['year'].value_counts().head(20).plot.bar()
```

as we seeen from above that in 2020 has release highest horror movies 

```python
sub_data=data[['genre_names','budget','title']].sort_values(by='budget',ascending=False).head(40)
plt.figure(figsize=(15,6))
sns.barplot(data=sub_data,x='title',y='budget')
plt.xticks(rotation = 90)
plt.show()
```

As we seen from above  that movie messenger and world war z  has highest budget 

```python
nd=pd.DataFrame(data.genre_names.value_counts().head(30)).reset_index()
plt.figure(figsize=(15,6))
sns.barplot(data=nd,x='index',y='genre_names')
plt.xticks(rotation = 90)
plt.show()
```

as we seen that their are  highly horror movies

```python
sub_df = data[['popularity','title','budget','vote_average']]
sub_df = sub_df[sub_df.budget != 0]
sub_df = sub_df.sort_values(by='popularity',ascending=False).head(20)
sub_df
ax1 = sns.set_style(style=None, rc=None )
fig, ax1 = plt.subplots(figsize=(20,6))
plt.xticks(rotation=45)
plt.title("Popularity x budget for each movie")

sns.lineplot(data = sub_df['popularity'], marker='o', sort = False, ax=ax1)
ax2 = ax1.twinx()
sns.barplot(data=sub_df,x = sub_df.title,y = sub_df.budget,ax=ax2,alpha=0.5,color='orange')
```

```python
#Cheaking gross total yearly
stop_word=set(STOPWORDS)
text=" ".join(data['genre_names'])
word_cloud=WordCloud(background_color='white',stopwords=stop_word,random_state=42,max_words=50).generate(text)
plt.figure(figsize=(15,10))
plt.imshow(word_cloud)
plt.axis('off')
plt.show()
```
