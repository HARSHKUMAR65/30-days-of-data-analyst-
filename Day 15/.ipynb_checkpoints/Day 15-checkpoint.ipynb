```python
#import libraries 
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import scipy.stats as  s
```

```python
## import libraries
gas=pd.read_csv("gasolinePricesInRomania.csv")
diseal=pd.read_csv("dieselPricesInRomania.csv")


## combine both data 
data=pd.merge(gas,diseal)
data.head()
```

Now Let us test hypothesis testing 

-->Ho => we have no significant relationship between features 
---------------------------------------------------------------------------------------------------------------------------
-->H1 =>We really have statisticaly relationship so we found pearson corelation and p-value for cheaking our hypothsis
------------------------------------------------------------------------------------------------------------------------------
1. if p-value < 0.05 we chose H1
2. if p-value > 0.05 we chose Ho

----------------------------------------------------------------------------------------------------------------------------

So now we found corelation cofficent useing scipy library

```python
res=s.pearsonr(data['gasoline price per liter in RON'],data['diesel price per liter in RON'])
```

```python
print(f"correlation is {res[0]} \n p-value is {res[1]}")
```

```python
plt.figure(figsize=(15,6))
plt.title('corelation cofficent')
sns.scatterplot(data=data,x='gasoline price per liter in RON',y='diesel price per liter in RON')
plt.show()
```

As we see we have very good corelation cofficent and  p-value < 0.5 so we chose H1 it is good statistically significant relationship between the feature 

# Now Let us see the change of price 

```python
plt.figure(figsize=(15,6))
plt.plot(data['date'], data['gasoline price per liter in RON'], color='green', label = 'diesel')
plt.plot(data['date'], data['diesel price per liter in RON'], color='red', label = 'gasoline')
plt.xticks(np.arange(0, len(data), 100))
plt.legend(loc='best', fontsize=12)
plt.xlabel('Dates', fontsize=14)
plt.ylabel('Price', fontsize=14)
plt.title('Change in energy prices from 2020 to 2023', fontsize = 14)
plt.show()
```
