```python
#import libraries 
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
```

```python
#import data
data=pd.read_csv("clean_dataa")
#first five rows 
data.head()
```

```python
# build kmean 
kmeans = KMeans(n_clusters=4, max_iter=50)
kmeans.fit(data[['incomeperperson','internetuserate','urbanrate']])
```

```python
#kmean
kmeans.labels_
```

```python
# Calculating sum of squared distance

ssd = []

for num_clusters in range(1,11):
    kmeans = KMeans(n_clusters=num_clusters, max_iter=50)
    kmeans.fit(data[['incomeperperson','internetuserate','urbanrate']])
    
    ssd.append(kmeans.inertia_)
```

the elbow curve

```python
plt.figure(figsize=(12,6))

plt.plot(range(1,11), ssd, linewidth=2, color="red", marker ="8")
plt.title("Elbow Curve\n=================================================================", fontsize=20, color="green")
plt.xlabel("K Value")
plt.xticks(np.arange(1,11,1))
plt.ylabel("SSD")

plt.show()
```

from above conclusion we  see that we take no of cluster is 3

```python
kmeans = KMeans(n_clusters=3, max_iter=50)
kmeans.fit(data[['incomeperperson','internetuserate','urbanrate']])
```

```python
kmeans.labels_
```

```python
data["Label"] = kmeans.labels_
```

```python
data
```

```python
# ScatterPlot when K=5

plt.figure(figsize=(10,6))

plt.title("Ploting the data into 5 clusters\n=================================================================", fontsize=20, color="green")
sns.scatterplot(data=data, x="country", y="internetuserate", hue="Label", s=60, palette=['green','orange','brown','blue'])
plt.show()
```

```python
# Getting the CustomerId for each group

cust1 = data[data.Label==0]
print("The number of country in 1st group = ", len(cust1))
print("The countries are - ", cust1.country.values)
print("============================================================================================\n")

cust2 = data[data.Label==1]
print("The number of country  in 2nd group = ", len(cust2))
print("The countries are - ", cust2.country.values)
print("============================================================================================\n")

cust3 = data[data.Label==2]
print("The number of country in 3rd group = ", len(cust3))
print("The countries are - ", cust3.country.values)
print("============================================================================================\n")
```
