```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
data=pd.read_csv("online_retail.csv")

data.head()
```

```python
##cheking null value 
data.isnull().sum()
```

```python
data.shape
```

```python
data.fillna(0,inplace=True)
```

```python
data
```

as we see there is no null value anymore so we need to  cheak d type 

```python
# cheakinng dtype 
data.info()
```

```python
#data clean and manuplation 
data['InvoiceDate']=pd.to_datetime(data['InvoiceDate'])
data['CustomerID']=data['CustomerID'].astype('int64')
data['total_price']=data['Quantity']*data['UnitPrice']
data['month']=data['InvoiceDate'].dt.month
data['year']=data['InvoiceDate'].dt.year
data['date']=data['InvoiceDate'].dt.date
```

```python
#sale per month 
plt.figure(figsize=(15,6))
sns.barplot(data=data,x='year',y='total_price')
plt.show()
```

```python
#sale per month 
plt.figure(figsize=(15,6))
sns.barplot(data=data,x='month',y='total_price',hue='year')
plt.show()
```

```python
data
```

```python
plt.figure(figsize=(15,6))
sns.countplot(data=data,x='Country')
plt.xticks(rotation=90)
plt.show()
```

```python
n_d=pd.DataFrame(data.groupby('Description')['total_price'].value_counts(ascending=False).drop(['Description']==0))
n_d.head(10).plot()
plt.xticks(rotation=90)
plt.show()
```
